```python
def mkdown_table_to_df(string):
    '''Takes a markdown table as a string and converts it into a dataframe'''
    import pandas as pd
    header = [c.strip() for c in string.split('\n')[0].split('|') if len(c) > 0]
    data = [[l.strip() for l in line.split('|') if len(l.strip()) > 0] for line in string.split('\n')[2:]][:-1]
    i = 0
    data_dict = {}

    for col in header:
        try:
            data_dict[col] = [line[i] for line in data]
            i += 1
        except IndexError:
            pass
        
    return pd.DataFrame(data_dict)
```

```python
def df_to_mkdown_table(df, to_print=True):
    '''Takes a pandas dataframe and converts it into a markdown table'''
    header = '| '+' | '.join(df.columns)+' | '
    divide = '|'+' -- |'*len(df.columns)
    data = ['| '+' | '.join(line)+' | ' for line in [r.values() for r in df.to_dict('records')]]
    if to_print:
        print(header)
        print(divide)
        for line in data:
            print(line)
    else:
        return [header]+[divide]+data
```
